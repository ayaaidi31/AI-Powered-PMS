## Docling Parsing & Chunking

```python
import os
import glob
from docling.document_converter import DocumentConverter
from langchain_text_splitters import RecursiveCharacterTextSplitter
from google.colab import userdata, drive

#  Mount Drive
drive.mount('/content/drive')

# 1. SETUP PATHS
PDF_FOLDER_PATH = '/content/drive/MyDrive/AWMF_Guidelines/'
pdf_files = glob.glob(os.path.join(PDF_FOLDER_PATH, "*.pdf"))

print(f"Found {len(pdf_files)} PDFs to process.")

# 2. INITIALIZE DOCLING CONVERTER
converter = DocumentConverter()

# 3. INITIALIZE MARKDOWN SPLITTER
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=923,
    chunk_overlap=200,
    separators=["\n\n", "\n", " ", ""]
)

all_docling_chunks = []

# 4. PROCESS EACH PDF
for i, pdf_path in enumerate(pdf_files):
    filename = os.path.basename(pdf_path)
    print(f"[{i+1}/{len(pdf_files)}] Parsing with Docling: {filename}...")

    try:
        # Let Docling's Vision AI read the layout and tables
        result = converter.convert(pdf_path)

        # Export the beautifully formatted result to Markdown
        md_text = result.document.export_to_markdown()

        # Chunk the Markdown text
        chunks = text_splitter.create_documents(
            texts=[md_text],
            metadatas=[{"source": filename, "parser": "docling"}]
        )
        all_docling_chunks.extend(chunks)

    except Exception as e:
        print(f"Error processing {filename}: {e}")

print(f"Docling extraction complete! Total chunks generated: {len(all_docling_chunks)}")
```

## Embed and Store in a Parallel Database

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_postgres import PGVector
from google.colab import userdata

# 1. INITIALIZE EMBEDDER
print("Loading BAAI/bge-m3 Embedder...")
bge_embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-m3",
    model_kwargs={'device': 'cpu'}
)

# 2. CONNECT TO NEON POSTGRES WITH A *NEW* COLLECTION NAME
NEON_CONNECTION_STRING = userdata.get('NEON_DATABASE_URL')
NEW_COLLECTION_NAME = "awmf_docling_bge" # New parallel database

print(f"Connecting to PGVector collection: {NEW_COLLECTION_NAME}...")
vector_store = PGVector(
    embeddings=bge_embeddings,
    collection_name=NEW_COLLECTION_NAME,
    connection=NEON_CONNECTION_STRING,
    use_jsonb=True
)

# 3. UPLOAD CHUNKS IN BATCHES
# Uploading in batches prevents memory crashes
batch_size = 200
total_chunks = len(all_docling_chunks)

print(f"Starting upload of {total_chunks} chunks...")

for i in range(0, total_chunks, batch_size):
    batch = all_docling_chunks[i : i + batch_size]
    vector_store.add_documents(batch)
    print(f"   Uploaded {min(i + batch_size, total_chunks)} / {total_chunks} chunks.")

print(f"SUCCESS! Parallel database '{NEW_COLLECTION_NAME}' is ready for the A/B test.")
```
